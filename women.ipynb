```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv("women.csv")
```

```python
df.head(10)
```

```python
df.info()
df.isnull().sum()
```

```python
# Rename columns for ease of use
df.columns = [
    'country',
    'security_index_2023',
    'danger_index_total_2019',
    'street_safety_2019',
    'intentional_homicide_2019',
    'non_partner_violence_2019',
    'intimate_partner_violence_2019',
    'legal_discrimination_2019',
    'gender_gap_2019',
    'gender_inequality_2019',
    'attitudes_toward_violence_2019'
]
```

```python
df = df.fillna(df.median(numeric_only = True))
```

```python
df.isnull().sum()
```

```python
df.dtypes
df.head()
```

```python
#looping through columns to plot histograms for each factor
for column in df.columns[1:]:
    plt.figure(figsize = (0,4))
    sns.histplot(df[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.show()
```

```python
# Exclude non-numeric columns (like 'country') for correlation calculation
numeric_data = df.select_dtypes(include=['float64', 'int64'])

# Calculate and plot the correlation matrix
plt.figure(figsize=(12, 8))
correlation_matrix = numeric_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot Street Safety by Country using the renamed column
plt.figure(figsize=(10, 6))
sns.barplot(x='street_safety_2019', y='country', data=df)
plt.title('Street Safety by Country')
plt.xlabel('Street Safety Score')
plt.ylabel('Country')
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Select only numeric columns for clustering, excluding 'country'
clustering_data = df.drop(columns=['country'])

# Initialize and fit the KMeans model
kmeans = KMeans(n_clusters=3, random_state=42)  # Set random_state for reproducibility
df['Cluster'] = kmeans.fit_predict(clustering_data)

# Visualize clusters using a scatter plot of two key variables
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='street_safety_2019', 
                y='intentional_homicide_2019', hue='Cluster', palette='viridis')
plt.title('Clustering Countries by Safety Indicators')
plt.xlabel('Street Safety Score')
plt.ylabel('Intentional Homicide Rate')
plt.show()
```
